```python
%matplotlib inline
import sys
print(sys.version)
import numpy as np
print(np.__version__)
import pandas as pd
print(pd.__version__)
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('raw_running_data.csv', parse_dates=['Date'])
```

```python
df.set_index('Date', inplace=True)
```

```python
df['day-of-week'] = df.index.map(lambda x: x.strftime('%A'))
```

```python
df['km'] = df.Miles.map(lambda x: x*1.60934)
```

```python
df['month'] = df.index.map(lambda x: x.strftime('%B'))
```

```python
# pay attention here, join happens on index
temp = pd.DataFrame(index=pd.date_range(df.index.min(), df.index.max()))
df = df.join(temp, how='outer')
```

```python
df.Miles.fillna(0, inplace=True)
```

```python
def get_total_seconds(raw_time):
    if str(raw_time) == 'nan': # is np.nan:
        return np.nan # if it's NaN, leave it
    hrs, mins, seconds = str(raw_time).split(':')
    seconds = int(seconds) + 60 * int(mins) + 60 * 60 * int(hrs)
    return seconds    
print(get_total_seconds(df.Time[0]))
# df.Time[0]
```

```python
df['seconds'] = df.Time.map(get_total_seconds)
```

```python
df['minutes'] = df.seconds.map(lambda x: x/60)
```

```python
df['month'].value_counts()
```

```python
mon = df.groupby('month')
# for name in mon.groups:
#     print(name)
# mon.get_group('August')
mon.agg(lambda x: x.mean())['Miles'].plot()
```
